## c_Print_formats

This program demonstrates various ways to use the print function in Python. It shows how to print variables and text together using different formatting techniques, including simple concatenation, the format() method, f-strings, and the % operator.

```python
# This program demonstrates various ways to use the print function in Python.
# It shows how to print variables and text together using different formatting techniques,
# including simple concatenation, the format() method, f-strings, and the % operator.

name = "Alice"
age = 30

# Simple printing
print("Name:", name, "Age:", age)

# Using string concatenation
print("Name: " + name + " Age: " + str(age))

# Using string formatting with .format()
print("Name: {} Age: {}".format(name, age))

# Using f-string (formatted string literals) - Recommended for Python 3.6+
print(f"Name: {name} Age: {age}")

# Using % operator (similar to printf in C)
print("Name: %s Age: %d" % (name, age))
```

## d_Add_list_of_numbers

This program asks the user to input 5 integers and stores them in a list. It calculates the total sum of these numbers and then prints the sum. Additionally, it prints out numbers from the list that are greater than 10.

```python
# -*- coding: utf-8 -*-
# This program asks the user to input 5 integers and stores them in a list. It calculates the total sum of these numbers
# and then prints the sum. Additionally, it prints out numbers from the list that are greater than 10.

# Create an empty list to store the numbers
numbers = []

# Ask the user to enter 5 integers
for i in range(5):
    number = int(input("Enter a number:"))
    numbers.append(number)

# Calculate the total sum of the numbers
total_sum = sum(numbers)
print(f"The total sum of the numbers is: {total_sum}")

# Print numbers greater than 10
print("Numbers greater than 10:")
for number in numbers:
    if number > 10:
        print(number)
```

## e_Multiplication_table_FOR

Program to generate a multiplication table for numbers 1 through 10 This example demonstrates the use of nested for loops to calculate and print the multiplication results in a structured format.

```python
# Program to generate a multiplication table for numbers 1 through 10
# This example demonstrates the use of nested for loops to calculate and print
# the multiplication results in a structured format.

# Outer loop for each number in the table
for i in range(1, 11):
    # Inner loop for multiplying the current number by each number 1 through 10
    for j in range(1, 11):
        # Calculate the product
        product = i * j
        # Print the formatted multiplication fact
        print(f"{i} * {j} = {product}")
    # Print a new line after each row of the table to separate the numbers
    print("-" * 20)
```

## f_Lists_example

This program showcases common list operations in Python, illustrating how to manipulate lists. Operations demonstrated include appending elements, inserting elements at specific positions, sorting the list, removing elements by value, popping elements by index, reversing the list, counting occurrences of an element, and finding an element's index.

```python
# This program showcases common list operations in Python, illustrating how to manipulate lists.
# Operations demonstrated include appending elements, inserting elements at specific positions,
# sorting the list, removing elements by value, popping elements by index, reversing the list,
# counting occurrences of an element, and finding an element's index.


# Define a list of strings
fruits = ["apple", "banana", "apple", "cherry"]

# Append an item to the list
fruits.append("orange")
print ("a) ",fruits)

# Insert an item at a specific index
fruits.insert(1, "blueberry")  # Insert "blueberry" at index 1
print ("b) ",fruits)

# Sort the list in alphabetical order
fruits.sort()
print ("c) ",fruits)

# Removing an element by value
fruits.remove("banana")
print ("d) ",fruits)

# Pop an element by index (remove and return the last element if no index is specified)
last_fruit = fruits.pop()
print ("e) ",fruits)

# Reverse the list
fruits.reverse()
print ("f) ",fruits)

# Get the number of occurrences of an element
count_apple = fruits.count("apple")

# Find the index of an element
index_of_cherry = fruits.index("cherry")

print("Modified list:", fruits)
print("Last fruit removed:", last_fruit)
print("Count of 'apple':", count_apple)
print("Index of 'cherry':", index_of_cherry)
```

## g_average_FUNCTION

This program calculates the average temperature from a list of temperatures. It demonstrates the use of functions, lists, and basic arithmetic operations in Python.

```python
# This program calculates the average temperature from a list of temperatures.
# It demonstrates the use of functions, lists, and basic arithmetic operations in Python.

def average_temperature(temp):
    """
    Calculate the average of the given list of temperatures.

    Parameters:
    temperatures (list): A list of temperatures (float or int values).

    Returns:
    float: The average temperature.
    """
    # Calculate the sum of all temperatures and divide by the length of the list
    return sum(temp) / len(temp)

# List of temperatures
temperatures = [22.0, 18.5, 25.3, 30.0, 16.2]

# Calculate the average temperature
avg_temp = average_temperature(temperatures)

# Print the average temperature
print ("Temperatures: ", temperatures)
print(f"The average temperature is: {avg_temp:.2f} degrees")
```

## h_ATM_example_IF_WHILE

This program simulates a simple ATM system in Python. Users start with a balance of $1000 and can withdraw money in multiples of $100. The program checks if the withdrawal amount is valid (a multiple of $100 and does not exceed the current balance). If the user attempts to withdraw more than their balance or an invalid amount, an error message is displayed. The loop continues until the user exits by entering '0'. This demonstrates basic use of control flow with while, if, elif, and else statements in Python.

```python
# Simple ATM Simulator
# This program simulates a simple ATM system in Python. Users start with a balance of $1000
# and can withdraw money in multiples of $100. The program checks if the withdrawal amount
# is valid (a multiple of $100 and does not exceed the current balance). If the user attempts
# to withdraw more than their balance or an invalid amount, an error message is displayed.
# The loop continues until the user exits by entering '0'. This demonstrates basic use of
# control flow with while, if, elif, and else statements in Python.

# Initial balance
balance = 1000

# Start of the program
print("Welcome to the ATM simulator. Your current balance is $1000")
print("Enter the amount to withdraw in multiples of $100, or '0' to exit.")

# Main loop
while True:
    # Ask the user for the amount to withdraw
    amount = int(input("Amount to withdraw: "))

    # Check for exit condition
    if amount == 0:
        print("Exiting. Thank you for using the ATM simulator.")
        break
    # Validate withdrawal amount
    elif amount % 100 == 0 and amount <= balance:
        balance -= amount  # Deduct the amount from balance
        print(f"${amount} withdrawn. New balance: ${balance}")
    else:
        # Handle errors: amount not in multiples of 100 or exceeds balance
        print("Invalid amount. Please enter a multiple of $100 not exceeding your balance.")
```

## i_Temp_conversion_FUNCTIONS

This program is designed to demonstrate the application of functions in Python for temperature conversion. It allows the user to input a temperature and choose whether they want to convert it from Celsius to Fahrenheit or from Fahrenheit to Celsius. The program defines two functions, one for each conversion direction, and uses conditional logic to execute the appropriate conversion based on user input. It showcases fundamental programming concepts such as function definition, arithmetic operations, conditional statements, and user input handling.

```python
# This program is designed to demonstrate the application of functions in Python for temperature conversion.
# It allows the user to input a temperature and choose whether they want to convert it from Celsius to Fahrenheit
# or from Fahrenheit to Celsius. The program defines two functions, one for each conversion direction, and uses
# conditional logic to execute the appropriate conversion based on user input. It showcases fundamental programming
# concepts such as function definition, arithmetic operations, conditional statements, and user input handling.

def convert_to_celsius(fahrenheit):
    #Converts Fahrenheit to Celsius.
    return (fahrenheit - 32) * 5/9

def convert_to_fahrenheit(celsius):
    #Converts Celsius to Fahrenheit.
    return (celsius * 9/5) + 32

# Main program
temp = float(input("Enter the temperature: "))
unit = input("Convert to (F)ahrenheit or (C)elsius? ")

if unit.upper() == 'F':
    print(f"{temp}C is {convert_to_fahrenheit(temp)}F")
elif unit.upper() == 'C':
    print(f"{temp}F is {convert_to_celsius(temp)}C")
else:
    print("Invalid unit.")
```

## j_Shopping_list

This program allows the user to create and manage a shopping list interactively. The user can add items to the shopping list by entering the names of the items one by one. To finish and see all the items added to the list, the user must enter '0'. The program then displays the complete shopping list before ending its execution. It is a simple but practical tool for keeping track of pending purchases.

```python
"""
Simplified Shopping List Program

Title: Interactive Shopping List

Description:
This program allows the user to create and manage a shopping list interactively.
The user can add items to the shopping list by entering the names of the items one by one.
To finish and view all the items added to the list, the user must enter '0'.
The program then displays the complete shopping list before ending its execution.
It is a simple but practical tool for keeping track of pending purchases.
"""

# Instruction for the user
print("Welcome to the shopping list program.")                                  # Welcomes the user to the program
print("Please enter the items you want to add to your list.")                   # Asks for item input
print("When you have finished, type '0' to end and show the complete list.\n")  # Instructions to finish the input

# Initialization of the shopping list
shopping_list = []                                                              # Creates an empty list to store the shopping items

# Loop to add items to the list
while True:
    # Asks the user for an item to add
    item = input("Enter the name of the item to add or '0' to finish: ")        # Gets the item name or '0' to end
    # Checks if the user wants to end
    if item == '0':
        break                                                                   # Exits the loop
    # Adds the item to the shopping list
    shopping_list.append(item)                                                  # Adds the item to the list

# Checks if the list is empty before displaying it
if shopping_list:
    # Displays all the items in the list
    print()
    print("Shopping List:")                                                     # Prints the heading for the shopping list
    for item in shopping_list:
        print(f"- {item}")                                                      # Prints each item in the list
else:
    # Informs the user that the list is empty
    print("No items were added to the list.")                                   # Message if no items were added
```

## k_simple_calculator

This program simulates a calculator that can perform basic arithmetic operations in sequence. The user enters a number, selects an arithmetic operation (+, -, /, x), and enters another number. The calculator executes the operation and waits for the next user input. This process is repeated until the user enters '=' to display the result and end the program.

```python
# Sequential Interactive Calculator

# This program simulates a calculator that can perform basic arithmetic operations sequentially.
# The user enters a number, selects an arithmetic operation (+, -, /, x), and enters another number.
# The calculator performs the operation and waits for the next user input.
# This process repeats until the user enters '=' to display the result and end the program.

# Ask the user for the first number
print("Sequential Interactive Calculator")
result = float(input("Enter the first number: "))                       # Reads the first number and converts it to float

# Start a loop to request additional operations and numbers
while True:
    # Ask the user for the operation to be performed
    operation = input("Enter the operation (+, -, /, x, =): ")          # Reads the operation

    # Check if the user wants to end and show the result
    if operation == "=":
        print("The result is:", result)                                 # Displays the final result
        break                                                           # Ends the loop and exits the program

    # Ask the user for the next number
    number = float(input("Enter the next number: "))                    # Reads the next number and converts it to float

    # Use match to determine the operation to be performed
    if operation == "+":
        result += number                                                # Adds the number to the result
    elif operation == "-":
        result -= number                                                # Subtracts the number from the result
    elif operation == "/":
        if number != 0:                                                 # Prevents division by zero
            result /= number
        else:
            print("Error: Division by zero. Please try again.")         # Error message for division by zero
            continue                                                    # Allows the user to correct their input without ending the program
    elif operation == "x":
        result *= number                                                # Multiplies the result by the number
    else:  # Captures any input that is not a valid operation
        print("Invalid operation. Please try again.")                   # Error message for invalid operation

# Note: This program does not handle errors for non-numeric input. It is assumed that the user inputs valid numbers.
```

## l_Guest_list

This program serves as a simple guest list manager for events. It allows users to interactively add, list, and remove guests from a list through a text-based menu interface. The key features include:
- **Adding Guests:** Users can continuously input names to add to the guest list until they choose to stop by pressing ENTER.
- **Listing Guests:** Displays all current guests in alphabetical order along with an index number for easy reference.
- **Removing Guests:** Users can delete a guest from the list by specifying the index number shown next to the guest's name.
- **Exiting the Program:** Users can exit the application at any time by selecting the exit option from the menu.

This program is designed to be user-friendly and is particularly useful for managing attendees in small to medium-sized events, providing basic functionalities needed to maintain an organized list of participants.

```python
# Guest List Management Program

def display_menu():
    """Prints the menu options on the console."""
    print("\nMenu:")                                        # Display a header for the menu
    print("A: Add - Add a new guest.")                      # Option to add guests
    print("L: List - Show the guest list.")                 # Option to display the list of guests
    print("R: Remove - Remove a guest from the list.")      # Option to remove a guest
    print("E: Exit - Exit the program.")                    # Option to exit the program
    return input("Enter an option (A, L, R, E): ").upper()  # Collect user input and convert it to uppercase

def add_guest():
    """Allows the user to add guest names until they press ENTER."""
    while True:                                             # Start a loop to repeatedly ask for guest names
        name = input("Enter the guest's name (press ENTER to finish): ")  # Get guest name from user
        if name == "":                                      # Check if the input is empty (ENTER was pressed)
            break                                           # Exit the loop if ENTER was pressed
        guest_list.append(name)                             # Add the guest name to the list
        print(name, "has been added to the guest list.")    # Confirm that the guest has been added

def list_guests():
    """Displays the guest list with indices."""
    print("\nGuest list:")                                  # Print a header for the guest list
    if len(guest_list) == 0:                                # Check if the guest list is empty
        print("The list is empty.")                         # Notify the user if the list is empty
        return                                              # Exit the function and return None
    guest_list.sort()                                       # Sort the guest list alphabetically
    index = 1                                               # Initialize index for guest listing
    for name in guest_list:                                 # Loop through each name in the sorted guest list
        print(index, ":", name)                             # Display each guest with their index
        index += 1                                          # Increment the index for the next guest

def remove_guest():
    """Allows the user to remove a guest from the list by the shown number."""
    if len(guest_list) == 0:                                # Check if the guest list is empty
        print("The list is empty.")                         # Notify the user if the list is empty
        return                                              # Exit the function and return None
    index_to_remove = int(input("Enter the number of the guest to remove: ")) - 1  # Ask for the index number to remove, adjusting for 0-based index
    if 0 <= index_to_remove < len(guest_list):              # Ensure the index is within the valid range
        guest_list.pop(index_to_remove)                     # Remove the guest from the list
        print("Guest removed from the list.")               # Confirm the removal
    else:
        print("Invalid number, please enter a valid number.")  # Handle out-of-range index values

# List to store the names of the guests
guest_list = []

# Main loop of the program
while True:
    option = display_menu()                         # Display the menu and get the user choice
    if option == "A":
        add_guest()                                 # Handle adding new guests
    elif option == "L":
        list_guests()                               # Handle displaying the guest list
    elif option == "R":
        remove_guest()                              # Handle removing a guest
    elif option == "E":
        print("End of the program.")                # Print exit message
        break                                       # Break the loop to exit the program
    else:
        print("Invalid option, please try again.")  # Handle invalid menu option entries
```

## m_Guest_list_plus


This program serves as a comprehensive guest list manager for events, allowing users to interactively manage guest information through a text-based menu interface. The key features of this program include:

- **Adding Guests:** Users can continuously input guest names, phone numbers, and emails to add to the guest list. The entry stops when the user presses ENTER without entering a name, allowing multiple guests to be added in a session.

- **Searching for Guests by Phone:** Allows users to search for a guest by their unique phone number and displays detailed information including name, phone, and email if found.

- **Deleting Guests:** Users can delete a guest from the list by entering the guest's phone number. This function enhances data management by allowing precise removal based on unique identifiers.

- **Listing All Guests:** Displays all current guests with their names, phone numbers, and emails. This function provides a complete overview of all the guests currently in the list.

- **Exiting the Program:** Users can exit the application at any time by selecting the exit option from the menu.

This system is designed to be user-friendly and is especially useful for managing attendees in various types of events. It provides essential functionalities needed to maintain a detailed and organized list of participants, supporting efficient event management by handling guest data interactively.

```python
# Guest List Management System


def add_guest():
    """Adds new guest to the guest list after getting user input."""
    while True:                                                             # Start a loop to repeatedly ask for guest names
        name = input("Enter the guest's name (press ENTER to finish): ")    # Ask for guest's name
        if name == "":                                                      # Check if the name is empty
            break                                                           # Exit the loop if no name is entered

        phone = input("Enter the guest's phone number (must be unique): ")  # Ask for guest's phone
        email = input("Enter the guest's email: ")                          # Ask for guest's email
        
        guest = {                                                           # Create a dictionary for the guest
            'Name': name,                                                   # Assign the guest's name
            'Phone': phone,                                                 # Assign the guest's phone
            'Email': email                                                  # Assign the guest's email
        }
        guest_list.append(guest)                                            # Add the guest to the global list
        print(f"Guest added with phone number: {phone}")                    # Confirm guest addition

def search_guest_by_phone():
    """Searches for a guest by phone number and displays their details if found."""
    search_phone = input("Enter the guest's phone number to search: ")      # Ask for the phone number to search
    found = False                                                           # Initialize flag to check if the guest is found
    for guest in guest_list:                                                # Iterate through the guest list
        if guest['Phone'] == search_phone:                                  # Check if the current guest's phone matches the search
            print("Guest Found:")                                           # Inform that the guest is found
            print(f"Name: {guest['Name']}, Phone: {guest['Phone']}, Email: {guest['Email']}")  # Display guest details
            found = True                                                    # Set found flag to True
            break                                                           # Break the loop after finding the guest

    if not found:                                                           # If no guest is found
        print("No guest found with that phone number.")                     # Inform the user

def delete_guest_by_phone():
    """Deletes a guest by phone number."""
    delete_phone = input("Enter the guest's phone number to delete: ")      # Ask for the phone number to delete
    index = 0                                                               # Initialize index to track the position
    found = False                                                           # Flag to check if the phone number is found

    while index < len(guest_list):                                          # Iterate through the list while index is less than its length
        if guest_list[index]['Phone'] == delete_phone:                      # Check if the current guest's phone matches
            guest_list.pop(index)                                           # Remove the guest from the list
            print(f"Guest with phone number {delete_phone} has been deleted.")  # Confirm deletion
            found = True                                                    # Set found flag to True
            break                                                           # Break the loop after deletion
        index += 1                                                          # Increment index if not found

    if not found:                                                           # If no guest is found
        print("No guest found with that phone number.")                     # Inform the user

def list_all_guests():
    """Lists all guests in the guest list."""
    if not guest_list:                                                      # Check if the guest list is empty
        print("No guests currently in the list.")                           # Inform the user if the list is empty
    else:
        print("Listing All Guests:")                                        # Print a header for the list
        for guest in guest_list:                                            # Iterate through the guest list
            print(f"Name: {guest['Name']}, Phone: {guest['Phone']}, Email: {guest['Email']}")  # Print each guest's details


# Global variable to store guest information
guest_list = []

# Main function to run the guest list management system
while True:
    print("\nGuest List Management System")                                 # Print the system menu header
    print("1. Add Guest")                                                   # Option to add a guest
    print("2. Search Guest by Phone")                                       # Option to search a guest by phone
    print("3. Delete Guest by Phone")                                       # Option to delete a guest by phone
    print("4. List All Guests")                                             # Option to list all guests
    print("5. Exit")                                                        # Option to exit the program

    choice = input("Enter your choice: ")                                   # Ask for user input
    if choice == '1':
        add_guest()                                                         # Call add_guest function
    elif choice == '2':
        search_guest_by_phone()                                             # Call search_guest_by_phone function
    elif choice == '3':
        delete_guest_by_phone()                                             # Call delete_guest_by_phone function
    elif choice == '4':
        list_all_guests()                                                   # Call list_all_guests function
    elif choice == '5':
        print("Exiting system.")                                            # Exit message
        break                                                               # Break the loop to exit the program
    else:
        print("Invalid choice. Please choose a valid option.")              # Handle invalid input

```
